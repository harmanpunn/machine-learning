```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sentences =[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
## One Hot Representation
one_hot_repr = [one_hot(words, voc_size) for words in sentences]
one_hot_repr
```

```python
## Word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sentence_length = 8
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sentence_length)
print(embedded_docs)
```

```python
## Feature representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sentence_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
